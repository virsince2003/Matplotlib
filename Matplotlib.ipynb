Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a Python library for data visualization. It is used to create high-quality plots and charts. The Pyplot module is a sublibrary within Matplotlib that provides a simplified interface for creating plots.  Five plots that can be created using the Pyplot module of Matplotlib are line plots, scatter plots, bar plots, histograms, and pie charts.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

The scatter() function is used to create scatter plots, which are effective for visualizing the relationship between two continuous variables.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
x_axis=plt.xlabel("x axis")
y_axis=plt.ylabel("y axis")
plt.title("scatter plot")
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots or charts in a structured layout. Each subplot represents a separate plot within the grid.

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([50, 20, 40, 20, 60, 70])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([10, 20, 30, 40, 50, 60])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Line Plot 1")

plt.subplot(2,2,2)
plt.plot(x1,y1)
plt.title("Line Plot 2")

plt.subplot(2,2,3)
plt.plot(x2,y2)
plt.title("Line Plot 3")

plt.subplot(2,2,4)
plt.plot(x3,y3)
plt.title("Line Plot 4")
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot, also known as a bar chart or bar graph, is a type of chart that uses rectangular bars to represent categorical data. The length of each bar corresponds to the value or frequency of the category being represented.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.ylabel("Companies")
plt.xlabel("Profit in revenue")
plt.barh(company,profit)
plt.title('Horizontal Bar Plot')
plt.show()

plt.xlabel("Companies")
plt.ylabel("Profit in revenue")
plt.bar(company,profit)
plt.title('Bar Plot')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of a dataset's distribution. It is used to display key measures such as the median, quartiles, and outliers, allowing for easy comparison between different groups or variables.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.figure(figsize=(6,10))
plt.boxplot([box1,box2])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis
fig, ax = plt.subplots()

# Create box plots
ax.boxplot([box1, box2])

# Set labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot')

# Display the plot
plt.show()
```
